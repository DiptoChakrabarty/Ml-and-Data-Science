# Proper Scaling of Data

```python
sns.countplot('Class',data=data)
plt.title('Class Distribution  Fraud:1/Not Fraud:0')  # Here there is a risk of overfitting the data towards not fraud
```

```python
fig,ax= plt.subplots(1,2,figsize=(15,4))
amt=data['Amount']
time=data['Time']

sns.distplot(amt,ax=ax[0],color='g')
ax[0].set_title('Distribution of Amount')

sns.distplot(time,ax=ax[1],color='r')
ax[1].set_title('Time variation')
```

```python
# Standard Scaling Time and Amount
std= StandardScaler()
rbs=RobustScaler()
data['Scaled_Amount']=rbs.fit_transform(data["Amount"].values.reshape(-1,1))
data['Scaled_Time']=rbs.fit_transform(data["Time"].values.reshape(-1,1))
data.drop(['Time','Amount'],axis=1,inplace=True)
```

```python
data.columns
```

```python
data.head()
```

```python
scaled_amount = data['Scaled_Amount']
scaled_time = data['Scaled_Time']

data.drop(['Scaled_Time', 'Scaled_Amount'], axis=1, inplace=True)
data.insert(0, 'Scaled_Amount', scaled_amount)
data.insert(1, 'Scaled_Time', scaled_time)
```

```python
data.head()
```

```python
cr =data.corr()
fig,ax = plt.subplots(figsize=(20,20))
sns.heatmap(cr,vmax=0.6,square=True,annot=True,cbar=True)
```

```python
X=data.drop(['Class'],axis=1)
y=data['Class']
```

```python
state=1
classifiers={
    "Isolation forest": IsolationForest(max_samples=len(X),contamination=(87/56874),random_state=state),
    "Local Outlier Factor": LocalOutlierFactor(n_neighbors=25,contamination=(87/56874))
}
```

```python
Fraud=data[data['Class']==1]
Fraud.head()
```

```python
n_outliers=len(Fraud)

for i ,(clf_name,clf) in enumerate(classifiers.items()):
    if clf_name=="Local Outlier Factor":
        pred=clf.fit_predict(X)
        scores_pred=clf.negative_outlier_factor_
    else:
        clf.fit(X)
        scores_pred=clf.decision_function(X)
        pred=clf.predict(X)
# Gives outliers as -1 and inliers as 1

    pred[pred==1]=0
    pred[pred==-1]=1
    
    n_error= (pred != y).sum()
    # Classification Metric
    print("{} {}".format(clf_name,n_error))
    print(accuracy_score(y,pred))
    print(classification_report(y,pred))
```
